```python
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt

# 1. Define specific phase angles (in degrees) and magnitudes
source_injections = {
    "Puttalam_Source": {"mag": 1.0, "angle_deg": 30.0},   # Example: 30 degrees
    "Barge_Source": {"mag": 1.0, "angle_deg": -45.0},     # Example: -45 degrees
    "Rantambe_Source": {"mag": 1.0, "angle_deg": 120.0}   # Example: 120 degrees
}

N = 165

I_matrix = np.eye(N, dtype=complex)
# Network Matrix M for the equation M @ a = 0 (homogeneous system)
# Assumes C_matrix_perfect and S_global are already defined
M = I_matrix - (C_matrix_perfect @ S_global) 

# Define source indices for the three sources in the system
source_indices = {
    "Puttalam_Source": 0,
    "Barge_Source": 96,
    "Rantambe_Source": 150
}

A_individual = {}
B_individual = {}
V_individual_complex = {}
V_individual_mag = {}

for src_name, k in source_indices.items():
    
    a_vec = np.zeros((N, 1), dtype=complex)
    
    # 2. Extract magnitude and convert degrees to radians
    mag = source_injections[src_name]["mag"]
    angle_rad = np.radians(source_injections[src_name]["angle_deg"])
    
    # 3. Apply Euler's formula to create the complex injection phasor
    a_vec[k, 0] = mag * np.exp(1j * angle_rad) 
    
    mask = np.arange(N) != k
    
    M_sub = M[np.ix_(mask, mask)]  
    M_col = M[mask, k].reshape(-1, 1)             
    
    # Solve for unknown incident waves
    rhs = -M_col * a_vec[k, 0]
    a_sub = la.solve(M_sub, rhs)
    
    # Reassemble 'a' and calculate 'b'
    a_vec[mask, 0] = a_sub.flatten()
    b_vec = S_global @ a_vec  
    
    V_individual_complex[src_name] = a_vec + b_vec
    A_individual[src_name] = a_vec.flatten()
    B_individual[src_name] = b_vec.flatten()
    V_individual_mag[src_name] = np.abs(V_individual_complex[src_name])

V_total_complex = sum(V_individual_complex.values())
V_total_mag = np.abs(V_total_complex)
a_total = sum(A_individual.values())
b_total = sum(B_individual.values())

nodes = np.arange(N)

plt.figure(figsize=(12, 6))
plt.plot(nodes, V_individual_mag["Puttalam_Source"], label='Puttalam Only (|a + b|)', alpha=0.6, linestyle='--')
plt.plot(nodes, V_individual_mag["Barge_Source"], label='Barge Only (|a + b|)', alpha=0.6, linestyle='--')
plt.plot(nodes, V_individual_mag["Rantambe_Source"], label='Rantambe Only (|a + b|)', alpha=0.6, linestyle='--')
plt.plot(nodes, V_total_mag, label='Base-Case Total (|a + b| summed)', color='black', linewidth=2)
plt.title('Base-Case Harmonic Voltage Envelope (250 Hz)', fontsize=14)
plt.xlabel('System Node Index', fontsize=12)
plt.ylabel('Voltage Magnitude (pu)', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend(fontsize=11)
plt.tight_layout()  
plt.show()
```

```python
import numpy as np


# --- System base ---
S_base   = 100e6      # VA
V_base   = 220e3      # V
Z_base   = V_base**2 / S_base  # 484 Ω
f1       = 50.0       # Hz fundamental
h        = 7          # harmonic order  ->  250 Hz
f_h      = h * f1     # analysis frequency
L_total  = 70.5       # km total line length

# -----------------------------------------------------------------------
# OHL per-unit-length parameters  (derived from PSSE, per circuit)
# -----------------------------------------------------------------------
R_pu = 0.00554
X_pu = 0.04457
B_pu = 0.13764

r_ohl = R_pu * Z_base / L_total                      # Ω/km
l_ohl = (X_pu * Z_base / (2*np.pi*f1)) / L_total     # H/km
c_ohl = (B_pu / Z_base / (2*np.pi*f1)) / L_total     # F/km
g_ohl = 0.0                                          # S/km  (lossless shunt)

# -----------------------------------------------------------------------
# UGC per-unit-length parameters  (220 kV XLPE cable, typical)
# -----------------------------------------------------------------------
r_ugc = 0.0277      # Ω/km  (Al conductor ~1200 mm²)
l_ugc = 0.38e-3     # H/km  (lower inductance than OHL)
c_ugc = 0.25e-6     # F/km  (high capacitance – key UGC feature)
g_ugc = 0.0         # S/km

# -----------------------------------------------------------------------
# Section lengths
# -----------------------------------------------------------------------
L_ohl1 = 25.25   # km  OHL  (Kotmale side)
L_ugc  = 20    # km  UGC  (middle)
L_ohl2 = 25.25   # km  OHL  (Biyagama side)
# Check: 25.25 + 20.0 + 25.25 = 70.5 ✓

# =============================================================================
# CORE FUNCTIONS
# =============================================================================

def compute_section(r, l, g, c, length_km, freq):
    
    omega  = 2 * np.pi * freq
    Z_pk   = r + 1j * omega * l
    Y_pk   = g + 1j * omega * c
    gamma  = np.sqrt(Z_pk * Y_pk)
    Z0     = np.sqrt(Z_pk / Y_pk)
    gl     = gamma * length_km
    A      = np.cosh(gl)
    B      = Z0 * np.sinh(gl)
    C      = np.sinh(gl) / Z0
    ABCD   = np.array([[A, B], [C, A]], dtype=complex)
    return gamma, Z0, ABCD


def ABCD_to_S(ABCD, Z_ref):
    
    A, B, C, D = ABCD[0,0], ABCD[0,1], ABCD[1,0], ABCD[1,1]
    Zr    = Z_ref
    denom = A + B/Zr + C*Zr + D
    S11   = (A  + B/Zr - C*Zr - D) / denom
    S12   = 2*(A*D - B*C)           / denom
    S21   = 2                        / denom
    S22   = (-A + B/Zr - C*Zr + D) / denom
    return np.array([[S11, S12], [S21, S22]], dtype=complex)


def parallel_ABCD(ABCD):
    
    return np.array([
        [ABCD[0,0], ABCD[0,1] / 2],
        [ABCD[1,0] * 2, ABCD[1,1]]
    ], dtype=complex)


# =============================================================================
# STEP 1 – Set reference impedance directly
# =============================================================================
Z_ref = 360.2607461227684 + 0j

# =============================================================================
# STEP 2 – Compute ABCD for each section of the new hybrid line (per circuit)
# =============================================================================
_, Z0_ohl, ABCD_ohl1  = compute_section(r_ohl, l_ohl, g_ohl, c_ohl, L_ohl1, f_h)
_, Z0_ugc, ABCD_ugc   = compute_section(r_ugc, l_ugc, g_ugc, c_ugc, L_ugc,  f_h)
_, _,      ABCD_ohl2  = compute_section(r_ohl, l_ohl, g_ohl, c_ohl, L_ohl2, f_h)

# =============================================================================
# STEP 3 – Cascade single circuit:  OHL1 → UGC → OHL2
# =============================================================================
ABCD_single = ABCD_ohl2 @ ABCD_ugc @ ABCD_ohl1   # right-to-left: OHL1 first

# =============================================================================
# STEP 4 – Combine two parallel circuits
# =============================================================================
ABCD_two_par = parallel_ABCD(ABCD_single)

# =============================================================================
# STEP 5 – Convert to S-parameters using Z_ref
# =============================================================================
line5_ugc_new = ABCD_to_S(ABCD_two_par, Z_ref)


line5_ohl_orig = np.array([
    [-0.20462834-0.33361586j,  0.78019381-0.48381641j],
    [ 0.78019381-0.48381641j, -0.20462834-0.33361586j]
], dtype=complex)

```

```python
# =========================================================
# 1. SETUP USING YOUR EXACT S-PARAMETER MATRICES
# =========================================================
# (Relies on S_global and C_matrix_perfect already in memory)
N = 165
K = 5  # Number of filters to place
epsilon = 1e-6

# Define the source injections based on your notebook's source indices
a_source_vector = np.zeros((N, 1), dtype=complex)
a_source_vector[0, 0]   = 0.8660 + 0.5000j  # Puttalam_Source
a_source_vector[96, 0]  = 0.7071 - 0.7071j  # Barge_Source
a_source_vector[150, 0] = -0.5000 + 0.8660j  # Rantambe_Source

# =========================================================
# 3. THE REVISED PHYSICS SOLVER (ACTIVE/DAMPED FILTER)
# =========================================================
def run_true_harmonic_solver(placed_filters_list):
    # 1. Start with the baseline S_global
    S_current = np.copy(S_global)
    
    # 2. Apply Filters (The Black Hole Method)
    for f_port in placed_filters_list:
        # By reducing the entire row and column for this port to near zero,
        # the component absorbs the wave perfectly. It does not reflect it back,
        # and it does not transmit it forward.
        S_current[f_port, :] *= 0.05
        S_current[:, f_port] *= 0.05
            
    # 3. Solve the Signal Flow Graph naturally
    I_matrix = np.eye(N, dtype=complex)
    M_current = I_matrix - (C_matrix_perfect @ S_current)
    
    # Calculate total incoming waves (a) and outgoing waves (b)
    a_total_curr = la.solve(M_current, a_source_vector)
    b_total_curr = S_current @ a_total_curr
    
    a_flat = a_total_curr.flatten()
    b_flat = b_total_curr.flatten()
    
    # 4. Extract Pseudo-T matrix for scoring
    current_T = np.zeros((N, N))
    for i in range(N):
        for j in range(N):
            if i != j and abs(a_flat[i]) > epsilon:
                current_T[i, j] = abs(b_flat[j]) / abs(a_flat[i])
                
    if np.max(current_T) > 0:
        current_T = current_T / np.max(current_T)
        
    return a_flat, b_flat, current_T

# =========================================================
# 3. RUN ITERATIVE FILTER PLACEMENT ALGORITHM
# =========================================================
print(f"\n--- Starting Iterative Filter Placement ({K} Filters) ---")

selected_filters = []
candidate_buses = list(range(N))

for k in range(K):
    # Get physical grid state 
    a_curr, b_curr, T_curr = run_true_harmonic_solver(selected_filters)
    influence = np.sum(T_curr, axis=1)
    
    scores = np.zeros(N)
    for i in candidate_buses:
        if abs(a_curr[i]) > epsilon: 
            scores[i] = abs(a_curr[i]) * abs(b_curr[i]) * influence[i]
            
    best_node = candidate_buses[np.argmax(scores[candidate_buses])]
    
    selected_filters.append(best_node)
    candidate_buses.remove(best_node)
    
    print(f"Filter {k+1} Placed at Node {best_node:3d} | Total Voltage |a+b| dropped to: {abs(a_curr[best_node] + b_curr[best_node]):.4f}")

# Extract Base Case and Final Case for Plotting
base_a, base_b, _ = run_true_harmonic_solver([])
final_a, final_b, _ = run_true_harmonic_solver(selected_filters)

# =========================================================
# 4. GENERATE PERFORMANCE BAR CHART
# =========================================================
print("\nAlgorithm complete. Generating True Network Performance Chart...")

# Calculate Total Voltage |a + b|
#mag_base = V_total_mag.flatten() 
mag_base = np.abs(base_a + base_b)  # Baseline total voltage magnitude (from previous superposition)
mag_final = np.abs(final_a + final_b)

fig, ax = plt.subplots(figsize=(20, 6))
x = np.arange(N)
width = 0.4  

ax.bar(x - width/2, mag_base, width, label='Base Total Voltage |a + b|', color='#e74c3c', alpha=0.8)
ax.bar(x + width/2, mag_final, width, label='Final Total Voltage |a + b|', color='#2ecc71', alpha=0.9)

ax.set_xlabel('Grid Node ID (0 to 164)', fontsize=12, fontweight='bold')
ax.set_ylabel('Total Voltage Magnitude |a + b|', fontsize=12, fontweight='bold')
ax.set_title(f'165-Node Performance ( S-Matrix Mutual Coupling included-After Adding UG)', fontsize=15, fontweight='bold')

ax.set_xticks(x[::5])
ax.set_xticklabels([str(i) for i in x[::5]], rotation=45)

for f_node in selected_filters:
    ax.get_xticklabels() 
    ax.text(f_node, mag_base[f_node] + 0.05, '★', ha='center', va='bottom', color='gold', fontsize=14, fontweight='bold')

ax.grid(axis='y', linestyle='--', alpha=0.5)
ax.legend(fontsize=12, loc='upper right')

plt.tight_layout()
plt.savefig('True_Mutual_Coupling_Performance.png', dpi=300)
plt.show()

print(f"\nFinal Optimal Filter Locations: {selected_filters}")
```
