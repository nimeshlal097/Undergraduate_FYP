```python
import numpy as np

# =============================================================================
# LINE 5: KOTMALE -> BIYAGAMA  |  UGC REPLACEMENT SCENARIO
# =============================================================================
# PSSE Data (50 Hz, 220 kV, 100 MVA base), two parallel circuits:
#   R = 0.00554 pu,  X = 0.04457 pu,  B = 0.13764 pu,  Length = 70.5 km
#
# Scenario: Replace middle 10 km with UGC:
#   [Kotmale] --OHL 30.25 km-- [UGC 10 km] --OHL 30.25 km-- [Biyagama]
#
# Analysis frequency: 250 Hz (5th harmonic, h=5 at 50 Hz base)
# Reference impedance: Z_ref = 360.2607461227684 Ω
# =============================================================================

# --- System base ---
S_base   = 100e6      # VA
V_base   = 220e3      # V
Z_base   = V_base**2 / S_base  # 484 Ω
f1       = 50.0       # Hz fundamental
h        = 5          # harmonic order  ->  250 Hz
f_h      = h * f1     # analysis frequency
L_total  = 70.5       # km total line length

# -----------------------------------------------------------------------
# OHL per-unit-length parameters  (derived from PSSE, per circuit)
# -----------------------------------------------------------------------
R_pu = 0.00554
X_pu = 0.04457
B_pu = 0.13764

r_ohl = R_pu * Z_base / L_total                      # Ω/km
l_ohl = (X_pu * Z_base / (2*np.pi*f1)) / L_total     # H/km
c_ohl = (B_pu / Z_base / (2*np.pi*f1)) / L_total     # F/km
g_ohl = 0.0                                          # S/km  (lossless shunt)

# -----------------------------------------------------------------------
# UGC per-unit-length parameters  (220 kV XLPE cable, typical)
# -----------------------------------------------------------------------
r_ugc = 0.0277      # Ω/km  (Al conductor ~1200 mm²)
l_ugc = 0.38e-3     # H/km  (lower inductance than OHL)
c_ugc = 0.25e-6     # F/km  (high capacitance – key UGC feature)
g_ugc = 0.0         # S/km

# -----------------------------------------------------------------------
# Section lengths
# -----------------------------------------------------------------------
L_ohl1 = 31.25   # km  OHL  (Kotmale side)
L_ugc  = 8.0    # km  UGC  (middle)
L_ohl2 = 31.25   # km  OHL  (Biyagama side)
# Check: 31.25 + 8.0 + 31.25 = 70.5 ✓

# =============================================================================
# CORE FUNCTIONS
# =============================================================================

def compute_section(r, l, g, c, length_km, freq):
    """
    Compute propagation constant γ, characteristic impedance Z0,
    and ABCD matrix for a uniform transmission line section.

    γ = sqrt((r + jωl)(g + jωc))
    Z0 = sqrt((r + jωl)/(g + jωc))
    ABCD: A=D=cosh(γl), B=Z0·sinh(γl), C=sinh(γl)/Z0
    """
    omega  = 2 * np.pi * freq
    Z_pk   = r + 1j * omega * l
    Y_pk   = g + 1j * omega * c
    gamma  = np.sqrt(Z_pk * Y_pk)
    Z0     = np.sqrt(Z_pk / Y_pk)
    gl     = gamma * length_km
    A      = np.cosh(gl)
    B      = Z0 * np.sinh(gl)
    C      = np.sinh(gl) / Z0
    ABCD   = np.array([[A, B], [C, A]], dtype=complex)
    return gamma, Z0, ABCD


def ABCD_to_S(ABCD, Z_ref):
    """
    Convert 2-port ABCD matrix to S-parameters with reference impedance Z_ref.

    S11 = (A + B/Zr - C·Zr - D) / Δ
    S12 = 2(AD - BC)             / Δ
    S21 = 2                      / Δ
    S22 = (-A + B/Zr - C·Zr + D)/ Δ
    Δ   = A + B/Zr + C·Zr + D
    """
    A, B, C, D = ABCD[0,0], ABCD[0,1], ABCD[1,0], ABCD[1,1]
    Zr    = Z_ref
    denom = A + B/Zr + C*Zr + D
    S11   = (A  + B/Zr - C*Zr - D) / denom
    S12   = 2*(A*D - B*C)           / denom
    S21   = 2                        / denom
    S22   = (-A + B/Zr - C*Zr + D) / denom
    return np.array([[S11, S12], [S21, S22]], dtype=complex)


def parallel_ABCD(ABCD):
    """
    Combine two IDENTICAL parallel circuits.
    B_parallel = B/2   (parallel halves series impedance)
    C_parallel = 2*C   (parallel doubles shunt admittance)
    A, D unchanged.
    """
    return np.array([
        [ABCD[0,0], ABCD[0,1] / 2],
        [ABCD[1,0] * 2, ABCD[1,1]]
    ], dtype=complex)


# =============================================================================
# STEP 1 – Set reference impedance directly
# =============================================================================
Z_ref = 360.2607461227684 + 0j

# =============================================================================
# STEP 2 – Compute ABCD for each section of the new hybrid line (per circuit)
# =============================================================================
_, Z0_ohl, ABCD_ohl1  = compute_section(r_ohl, l_ohl, g_ohl, c_ohl, L_ohl1, f_h)
_, Z0_ugc, ABCD_ugc   = compute_section(r_ugc, l_ugc, g_ugc, c_ugc, L_ugc,  f_h)
_, _,      ABCD_ohl2  = compute_section(r_ohl, l_ohl, g_ohl, c_ohl, L_ohl2, f_h)

# =============================================================================
# STEP 3 – Cascade single circuit:  OHL1 → UGC → OHL2
# =============================================================================
ABCD_single = ABCD_ohl2 @ ABCD_ugc @ ABCD_ohl1   # right-to-left: OHL1 first

# =============================================================================
# STEP 4 – Combine two parallel circuits
# =============================================================================
ABCD_two_par = parallel_ABCD(ABCD_single)

# =============================================================================
# STEP 5 – Convert to S-parameters using Z_ref
# =============================================================================
line5_ugc_new = ABCD_to_S(ABCD_two_par, Z_ref)

# =============================================================================
# RESULTS
# =============================================================================
print("=" * 68)
print("LINE 5 UGC SCENARIO  –  S-MATRIX RESULT")
print("=" * 68)
print(f"  Reference impedance  Z_ref = {Z_ref.real:.4f} Ω")
print(f"  OHL char. impedance  Z0_OHL = {Z0_ohl:.4f} Ω  (|Z0|={abs(Z0_ohl):.2f} Ω)")
print(f"  UGC char. impedance  Z0_UGC = {Z0_ugc:.4f} Ω  (|Z0|={abs(Z0_ugc):.2f} Ω)")
print(f"  Impedance mismatch ratio: {abs(Z0_ohl)/abs(Z0_ugc):.2f}x  "
      f"→ drives reflections at OHL-UGC junctions")
print()

line5_ohl_orig = np.array([
    [-0.20462834-0.33361586j,  0.78019381-0.48381641j],
    [ 0.78019381-0.48381641j, -0.20462834-0.33361586j]
], dtype=complex)

print(f"{'':>35} {'|S11|':>8} {'|S12|':>8}")
print(f"  line5_ohl  (original, pure OHL):   "
      f"{abs(line5_ohl_orig[0,0]):8.4f} {abs(line5_ohl_orig[0,1]):8.4f}")
print(f"  line5_ugc_new (OHL+UGC+OHL):       "
      f"{abs(line5_ugc_new[0,0]):8.4f} {abs(line5_ugc_new[0,1]):8.4f}")
d11 = abs(line5_ugc_new[0,0]) - abs(line5_ohl_orig[0,0])
d12 = abs(line5_ugc_new[0,1]) - abs(line5_ohl_orig[0,1])
print(f"  Δ (UGC impact):                    {d11:+8.4f} {d12:+8.4f}")
print()
print("  Physical interpretation:")
print(f"    Reflection |S11| increases by {d11:+.4f}  "
      f"({'↑ MORE' if d11>0 else '↓ less'} harmonic reflected back to Kotmale)")
print(f"    Transmission |S12| changes by {d12:+.4f}  "
      f"({'↓ LESS' if d12<0 else '↑ more'} harmonic propagates to Biyagama)")
print()

print("=" * 68)
print("USE IN YOUR CODE:")
print("=" * 68)
print()
print("# --- Line 5 Scenarios ---")
print()
print("line5_ohl = np.array([")
print(f"    [{line5_ohl_orig[0,0]:.8f}, {line5_ohl_orig[0,1]:.8f}],")
print(f"    [{line5_ohl_orig[1,0]:.8f}, {line5_ohl_orig[1,1]:.8f}]")
print("], dtype=complex)")
print()
print("# New scenario: 30.25 km OHL + 10 km UGC + 30.25 km OHL (2 parallel circuits)")
print(f"# Parameters: Z_ref={Z_ref.real:.4f}Ω, r_ugc=0.0277Ω/km, l_ugc=0.38mH/km, c_ugc=0.25μF/km")
print("line5_ugc_new = np.array([")
print(f"    [{line5_ugc_new[0,0]:.8f}, {line5_ugc_new[0,1]:.8f}],")
print(f"    [{line5_ugc_new[1,0]:.8f}, {line5_ugc_new[1,1]:.8f}]")
print("], dtype=complex)")
print()
print("# --- To run the UGC scenario, replace Line 5 in line_matrices ---")
print("line_matrices_ugc = dict(line_matrices)           # copy original")
print('line_matrices_ugc["Line 5"] = line5_ugc_new      # swap in UGC version')
print()
print("# Then rebuild S_global with line_matrices_ugc instead of line_matrices")
print("# and rerun the M matrix solve.")

# --- Symmetry & passivity check ---
print()
print("Checks:")
print(f"  Symmetric S12==S21:  {np.allclose(line5_ugc_new[0,1], line5_ugc_new[1,0])}")
print(f"  S11==S22:            {np.allclose(line5_ugc_new[0,0], line5_ugc_new[1,1])}")
print(f"  |S11|²+|S21|² = {abs(line5_ugc_new[0,0])**2 + abs(line5_ugc_new[1,0])**2:.6f}  (≤1 passive ✓)")
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

print("Generating CLEAN Before vs After UGC Heatmap...")

# ==============================
# 1. HELPER FUNCTIONS
# ==============================

def extract_node_values(V_data):
    values = []
    for node in node_list:
        ports = network_port_mapping[node]
        port_vals = [np.abs(V_data[p][0]) for p in ports]
        values.append(max(port_vals))
    return np.array(values)


def clip_values(vals, percentile=95):
    limit = np.percentile(vals, percentile)
    return np.clip(vals, None, limit)


# ==============================
# 2. EXTRACT DATA
# ==============================

node_list = list(G.nodes())

before_vals = extract_node_values(V_total_mag)
after_vals  = extract_node_values(V_total_mag_new)

# ------------------------------
# Clip extreme outliers
# ------------------------------
before_vals = clip_values(before_vals)
after_vals  = clip_values(after_vals)

# ------------------------------
# Log scaling (fix visibility)
# ------------------------------
before_log = np.log10(before_vals + 1e-6)
after_log  = np.log10(after_vals + 1e-6)

# ------------------------------
# Delta calculation (IMPORTANT)
# ------------------------------
delta_vals = after_vals - before_vals
max_abs_delta = np.max(np.abs(delta_vals))


# ==============================
# 3. GRAPH LAYOUT
# ==============================

pos = nx.kamada_kawai_layout(G)


# ==============================
# 4. PLOTTING
# ==============================

fig, axes = plt.subplots(1, 3, figsize=(30, 10))

cmap_main = plt.cm.inferno
cmap_delta = plt.cm.seismic

# ------------------------------
# BEFORE
# ------------------------------
nodes1 = nx.draw_networkx_nodes(
    G, pos, ax=axes[0],
    nodelist=node_list,
    node_color=before_log,
    cmap=cmap_main,
    node_size=1400,
    edgecolors='white'
)

nx.draw_networkx_edges(G, pos, ax=axes[0], alpha=0.3)
nx.draw_networkx_labels(G, pos, ax=axes[0], font_size=9)

axes[0].set_title("Before UGC (Log Scale)", fontsize=16, fontweight='bold')
axes[0].axis('off')


# ------------------------------
# AFTER
# ------------------------------
nodes2 = nx.draw_networkx_nodes(
    G, pos, ax=axes[1],
    nodelist=node_list,
    node_color=after_log,
    cmap=cmap_main,
    node_size=1400,
    edgecolors='white'
)

nx.draw_networkx_edges(G, pos, ax=axes[1], alpha=0.3)
nx.draw_networkx_labels(G, pos, ax=axes[1], font_size=9)

axes[1].set_title("After UGC (Log Scale)", fontsize=16, fontweight='bold')
axes[1].axis('off')


# ------------------------------
# DELTA (CRITICAL VIEW)
# ------------------------------
nodes3 = nx.draw_networkx_nodes(
    G, pos, ax=axes[2],
    nodelist=node_list,
    node_color=delta_vals,
    cmap=cmap_delta,
    vmin=-max_abs_delta,
    vmax=+max_abs_delta,
    node_size=1400,
    edgecolors='white'
)

nx.draw_networkx_edges(G, pos, ax=axes[2], alpha=0.3)
nx.draw_networkx_labels(G, pos, ax=axes[2], font_size=9)

axes[2].set_title("Impact (After - Before)", fontsize=16, fontweight='bold')
axes[2].axis('off')


# ==============================
# 5. COLORBARS
# ==============================

cbar1 = fig.colorbar(nodes1, ax=axes[:2], shrink=0.7)
cbar1.set_label('Log10(|V| pu)', fontsize=12)

cbar2 = fig.colorbar(nodes3, ax=axes[2], shrink=0.7)
cbar2.set_label('Δ|V| pu', fontsize=12)


# ==============================
# 6. HIGHLIGHT CRITICAL NODES
# ==============================

threshold = np.percentile(after_vals, 90)

critical_nodes = {
    node: node for node, val in zip(node_list, after_vals)
    if val >= threshold
}

nx.draw_networkx_labels(
    G, pos,
    labels=critical_nodes,
    font_size=10,
    font_color='black',
    font_weight='bold'
)


# ==============================
# 7. FINAL TOUCH
# ==============================

plt.tight_layout()
plt.show()


# ==============================
# 8. PRINT TOP WORST NODES
# ==============================

print("\nTop 10 Worst Nodes After UGC:\n")

sorted_nodes = sorted(
    zip(node_list, after_vals),
    key=lambda x: x[1],
    reverse=True
)

for i, (node, val) in enumerate(sorted_nodes[:10], 1):
    print(f"{i}. {node}: {val:.4f} pu")
```

```python
import json

def convert_ipynb_to_py(notebook_path, output_path):
    with open(notebook_path, 'r', encoding='utf-8') as f:
        nb = json.load(f)

    with open(output_path, 'w', encoding='utf-8') as out_f:
        for cell in nb.get('cells', []):
            if cell['cell_type'] == 'code':
                # Write code directly
                source_lines = cell.get('source', [])
                for line in source_lines:
                    out_f.write(line)
                out_f.write('\n\n')

            elif cell['cell_type'] == 'markdown':
                # Convert Markdown to Python comments
                source_lines = cell.get('source', [])
                for line in source_lines:
                    out_f.write(f"# {line}")
                out_f.write('\n\n')

    print(f"Successfully converted '{notebook_path}' to '{output_path}'")

# Run the conversion
convert_ipynb_to_py('S.ipynb', 'S.py')
```
